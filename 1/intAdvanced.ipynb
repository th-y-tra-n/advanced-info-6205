```python
############################################################
# Int.py
# Implements Int object
###########################################################

###########################################################
#  class  Int
###########################################################
class Int:
    def __init__(self, n: "Python int" = 0):
        # NOTHING CAM BE CHANGED IN THIS ROUTINE
        # ONLY DATA STRUCTURE ALLOWED
        # self._positive
        # self._a
        self._positive = True
        if n < 0:
            self._positive = False
        self._a = self.build(n)

    def _get_key(self)->'List of int':
        return self._a

    #############################
    # WRITE All public functions BELOW
    # YOU CAN HAVE ANY NUMBER OF PRIVATE FUNCTIONS YOU WANT
    #############################

    # To print the number
    def __str__(self):
        # String representation for print() and informal display
        return ('-' if not self._positive else '') + ''.join(map(str, self._a))

    def __repr__(self):
        # Official string representation, useful for debugging
        return f"Int({self.__str__()})"
    
    def __len__(self) -> "Python int size of the number":
        return len(self._a)
    
    def __getitem__(self, i: "Python int") -> "Python int":
        return self._a[i]
    
    def __setitem__(self, i: "Python int", v: "Python int") -> "None":
        self._a[i] = v

    # Operator Overloading
    def __add__(self, other: "Int") -> "Int":
        if not isinstance(other, Int):
            raise TypeError("Expected Int object")
        num1 = int(''.join(map(str, self._a)))
        num2 = int(''.join(map(str, other._a)))
        if not self._positive:
            num1 = -num1
        if not other._positive:
            num2 = -num2
        return Int(num1 + num2)
    
    def __sub__(self, other: "Int") -> "Int":
        if not isinstance(other, Int):
            raise TypeError("Expected Int object")
        num1 = int(''.join(map(str, self._a)))
        num2 = int(''.join(map(str, other._a)))
        if not self._positive:
            num1 = -num1
        if not other._positive:
            num2 = -num2
        return Int(num1 - num2)
    
    def __mul__(self, other: "Int") -> "Int":
        if not isinstance(other, Int):
            raise TypeError("Expected Int object")
        num1 = int(''.join(map(str, self._a)))
        num2 = int(''.join(map(str, other._a)))
        if not self._positive:
            num1 = -num1
        if not other._positive:
            num2 = -num2
        return Int(num1 * num2)
    
    def __truediv__(self, other: "Int") -> "Int":
        if not isinstance(other, Int):
            raise TypeError("Expected Int object")
        num1 = int(''.join(map(str, self._a)))
        num2 = int(''.join(map(str, other._a)))
        if not self._positive:
            num1 = -num1
        if not other._positive:
            num2 = -num2
        return Int(num1 // num2)
    
    # Comparison Operators
    def __lt__(self, other: "Int") -> "bool":
        if not isinstance(other, Int):
            raise TypeError("Expected Int object")
        num1 = int(''.join(map(str, self._a)))
        num2 = int(''.join(map(str, other._a)))
        if not self._positive:
            num1 = -num1
        if not other._positive:
            num2 = -num2
        return num1 < num2
    
    def __le__(self, other: "Int") -> "bool":
        if not isinstance(other, Int):
            raise TypeError("Expected Int object")
        num1 = int(''.join(map(str, self._a)))
        num2 = int(''.join(map(str, other._a)))
        if not self._positive:
            num1 = -num1
        if not other._positive:
            num2 = -num2
        return num1 <= num2
    
    def __eq__(self, other: "Int") -> "bool":
        if not isinstance(other, Int):
            raise TypeError("Expected Int object")
        num1 = int(''.join(map(str, self._a)))
        num2 = int(''.join(map(str, other._a)))
        if not self._positive:
            num1 = -num1
        if not other._positive:
            num2 = -num2
        return num1 == num2
    
    def __ne__(self, other: "Int") -> "bool":
        if not isinstance(other, Int):
            raise TypeError("Expected Int object")
        num1 = int(''.join(map(str, self._a)))
        num2 = int(''.join(map(str, other._a)))
        if not self._positive:
            num1 = -num1
        if not other._positive:
            num2 = -num2
        return num1 != num2
    
    # Hashing
    def __hash__(self) -> "int":
        return hash(tuple(self._a))

    #############################
    # -1986 is returned as [1, 9, 8, 6]
    # 1986  is returned as [1, 9, 8, 6]
    # -100  is returned as [1, 0, 0]
    # -0    is returned as [0]
    # TIME:O(log to base 10 of n)
    # SPACE:O(log to base 10 of n)
    #############################
    def build(self, n: "Python int") -> "list of int":
        # COnvert int to absolute value 
        abs_n = abs(n)
        print("abs_n: ", abs_n)
        # Convert the absolute integer to string, then to list
        self._a = [int(d) for d in str(abs_n)]
        print("self.a: ", self._a)
        return self._a   
        

    #############################
    # -1986 is stored as 1 9 8 6
    # 0 1 2 3
    # 1 9 8 6
    # return int value
    # TIME:O(log to base 10 of n)
    # SPACE:O(log to base 10 of n)
    #############################
    def int(self) -> "Python integer":
        #print("Remove this line and write code")
        # return an integer value
        # if the number is negative, return the negative value of the list
        # if the number is positive, return the positive value of the list
        # if the number is zero, return 0
        result = int (''.join(str(i) for i in self._a))
        if self._positive == False:
            result = -result
        return result


    
    ##############################################################
    # WRITE All private functions BELOW
    # YOU CAN HAVE ANY NUMBER OF PRIVATE FUNCTIONS YOU WANT
    ##############################################################
```
